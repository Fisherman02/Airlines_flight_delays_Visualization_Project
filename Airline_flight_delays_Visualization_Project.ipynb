```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
flight_df = pd.read_csv("C:/WALTER/Career/Projects/Projects/Covid_19_Data/New folder/flight_delays.csv", index_col = 'Month')
```

```python
flight_df.info()
flight_df.shape
```

```python
flight_df.sample(7)
```

```python
# Visualization of AA airline's Annual delay time

fig = plt.figure(figsize = (10,6))
ax = sns.barplot(x = flight_df.index, y = flight_df['AA'])
plt.title('Average Annual Flight delays at AA Airline')
plt.ylabel('Average Delay(Mins)')
plt.xlabel('Month')
plt.show()
```

```python
# Visualization of F9 Airline's annual delay time

fig = plt.figure(figsize = (12,8))
ax = sns.barplot(x = flight_df.index, y = flight_df['F9'])
plt.title('Average Annual Flight delays at F9 Airline')
plt.ylabel('Average Delay(Mins)')
plt.xlabel('Month')
plt.show()
```

```python
# Average Arrival Delay for Each Airline, by Month

fig = plt.figure(figsize = (10,8))
ax = sns.heatmap(data = flight_df, annot = True)
plt.title("Average Arrival Delay for Each Airline, by Month", size = 22)
plt.xlabel('Airline')
plt.ylabel('Monthly Average Flight Delays (mins)')                                           
plt.show()
                                       
                                       
                                       
```

```python
# Statistical Aggregate of Various Airlines flight delays

flight_df.describe().round(2)
```

```python
# Sum of Monthly Maximum delay time in a year 

highest_delays = flight_df.groupby('Month').max().sum()
max_annual_delay = highest_delays.sort_values(ascending = True).round(2)
max_annual_delay
```

```python
fig = plt.figure(figsize = (12,6))
ax = sns.lineplot(x = max_annual_delay.index, y = max_annual_delay.iloc[0:], color = 'green')
plt.title('Sum of Monthly Maximum delay time in a year', size = 20)
plt.xlabel('Airlines', size = 12)
plt.ylabel('Sum of Monthly Maximum delay time', size = 12)
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
ax = sns.barplot(x = max_annual_delay.index, y = max_annual_delay.iloc[0:])
plt.title('Sum of Monthly Maximum delay time in a year', size = 20)
plt.xlabel('Airlines', size = 12)
plt.ylabel('Sum of Monthly Maximum delay time', size = 12)
plt.show()
```
